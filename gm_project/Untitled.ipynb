```python
from numpy import dot
from numpy.linalg import norm
```

```python
def find_sim(X, j):
    cos_sim = []
    for i in X:
        cos_sim.append(dot(i, X[j])/(norm(i)*norm(X[j])))
    return cos_sim
```

```python
cos = find_sim(X_c, 0)
```
